Cleaning Data (Twice)



```python
combined = [data]

age_mapping = {"0-18":1, "18-30":2, "30-45":3, "45-60":4, "60-75":5, "75+":6}

for dataset in combined:
    dataset['Age'] = dataset['Age'].map(age_mapping)
    dataset['Age'] = dataset['Age'].fillna(0).astype(int)

data
```

```python
gender_mapping = {"Erkek":1, "Kadın":2}
data['Sex'] = data['Sex'].map(gender_mapping)
data
```

```python

combined = [data]
city_mapping = {"Marmara" : 1, "Ege":2, "Karadeniz":3, "Akdeniz":4, "İç Anadolu":5, "Doğu Anadolu":6, "Güneydoğu": 7}

for dataset in combined:
  dataset['Region'] = dataset['Region'].map(city_mapping)
  dataset['Region'] = dataset['Region'].fillna(0)
data
```

```python
combined = [data]
education_mapping = {"İlkokul": 1, "Ortaokul": 2, "Lise": 3, "Ön Lisans": 4, "Lisans": 5, "Lisans Üstü": 6}
for dataset in combined:
    dataset['Education'] = dataset['Education'].map(education_mapping)
    dataset['Education'] = dataset['Education'].fillna(0)

data.head()
```

```python
question_mapping = {"Evet": 1, "Hayır": 0}
for i in range(4,14):
    data[new_column_name[i]] = data[new_column_name[i]].map(question_mapping)

data.head()
```

```python
combine = [data]
opinion_mapping = {"AKP": 0, "CHP": 1, "IYI PARTI": 3}
for dataset in combine:
    dataset['parties'] = dataset['parties'].map(opinion_mapping)
    dataset['parties'] = dataset['parties'].fillna(0)

data.head()
```

##Modelling

```python
k = 14
corrmat = data.corr()
cols = corrmat.nlargest(k,'parties')['parties'].index
f, ax = plt.subplots(figsize=(10,5))
cm = np.corrcoef(data[cols].values.T)
sns.set(font_scale=1)
hm = sns.heatmap(cm, cbar=True, annot=True, square=True, fmt='.2f', annot_kws={'size':10}, yticklabels=new_column_name, xticklabels=new_column_name)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.tree import DecisionTreeClassifier, export_graphviz
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
```

```python
predictors = data.drop(['parties'], axis=1)
target = data["parties"]
x_train, x_test, y_train, y_test = train_test_split(predictors, target, test_size = 0.5, random_state=0)
```

```python
DT = DecisionTreeClassifier(max_depth=6)
DT.fit(x_train, y_train)
y_pred = DT.predict(x_test)
acc_DT = round(accuracy_score(y_pred, y_test) * 100, 2)
print(acc_DT)
```

```python
from re import M
export_graphviz(DT, feature_names=new_column_name, out_file='AKPvsCHPvsIYI.dot',
                filled=True,rounded=True)
```

```python
from sklearn.ensemble import RandomForestClassifier

randomforest = RandomForestClassifier()
randomforest.fit(x_train, y_train)
y_pred = randomforest.predict(x_test)
acc_randomforest = round(accuracy_score(y_pred, y_test) * 100, 2)
print(acc_randomforest)
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier()
knn.fit(x_train, y_train)
y_pred = knn.predict(x_test)
acc_knn = round(accuracy_score(y_pred, y_test) * 100, 2)
print(acc_knn)
```

```python
from sklearn.linear_model import LogisticRegression

logreg = LogisticRegression()
logreg.fit(x_train, y_train)
y_pred = logreg.predict(x_test)
acc_logreg = round(accuracy_score(y_pred, y_test) * 100, 2)
print(acc_logreg)
```

```python
from sklearn.svm import SVC

svm = SVC()
svm.fit(x_train, y_train)
y_pred = svm.predict(x_test)
acc_svm = round(accuracy_score(y_pred, y_test) * 100, 2)
print(acc_svm)
```

##Model Scores

```python
models = pd.DataFrame({
    'Model': ['SVC',
              'Random Forest', 'KNeighbors',
              'Decision Tree', 'Logistic Regression'],
    'Score': [ acc_svm,
              acc_randomforest, acc_knn, acc_DT, acc_logreg
             ]})
models.sort_values(by='Score', ascending=False)
```
